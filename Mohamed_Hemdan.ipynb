# Salt And Pepper Noise

```python
img_grayscale_noise = cv.imread('6.jpg',0)
img_grayscale_noise = img_grayscale_noise/255 # normalization 

x,y = img_grayscale_noise.shape
g = np.zeros((x,y), dtype=np.float32)
# salt and pepper amount
pepper = 0.1
salt = 0.95

# create salt and peper noise image    
for i in range(x):
    for j in range(y):
        rdn = np.random.random()
        if rdn < pepper:
            g[i][j] = 0
        elif rdn > salt:
            g[i][j] = 1
        else:
            g[i][j] = img_grayscale_noise[i][j]
            

img_noise = g
```

```python
# # Save image 
# cv.imwrite('5percent.jpg', img_as_ubyte(g))
```

# Mean filter (average)

```python
#  It removes the high-frequency content from the image. 
# It is also used to blur an image
## denoise image
m = 5
n = 5
denoise_mean = cv.blur(img_noise, (m,n))
```

# Median filter

```python
#  nonlinear filtering
# replaced by the median value of the neighboring pixel
img_noise_median = np.clip(img_noise, -1, 1) #pixel value range
img_noise_median = img_as_ubyte(img_noise_median) #convert to uint8
denoise_median = cv.medianBlur(img_noise_median, 5)
```
